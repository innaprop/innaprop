```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import numpy as np
import seaborn as sns
import glob
import sys
import os
import torch
sns.set(style="whitegrid")
```

```python
path = 'gpt2-mini'
all_files = glob.glob(path + '/*.csv')
df = pd.concat((pd.read_csv(f, sep=';') for f in all_files), ignore_index=True)
selected_values = ['AdamW',
                    'INNAprop, ($\\alpha$, $\\beta$) = (0.1, 0.9)'
]
mask = df['optimizer'].isin(selected_values)
df = df[mask]

palette = sns.color_palette("hls", len(df['optimizer'].unique()))

fig, ax = plt.subplots(1, 1, figsize=(5, 3))

df.sort_values(by=['optimizer', 'step'], inplace=True)
df['min_cumulative_validation_loss'] = df.groupby('optimizer')['validation_loss'].cummin()

# Plotting
sns.lineplot(data=df, x='step', y='min_cumulative_validation_loss', hue='optimizer', style='optimizer',
             dashes=False, palette=palette, ax=ax, legend=False)

ax.set_xlabel('Number of steps')
ax.set_ylabel('Validation loss')

ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x/1000)}K'))

ax.set_ylim(3.4, 4.5)

legend_handles = [mlines.Line2D([], [], color=palette[i], markersize=10, label=name)
                  for i, name in enumerate(df['optimizer'].unique())]
fig.legend(handles=legend_handles, loc='lower center', bbox_to_anchor=(0.7, 0.7))

plt.tight_layout()
plt.subplots_adjust(bottom=0.2)
plt.savefig('gpt2_mini.pdf', bbox_inches='tight')
plt.show()
```
