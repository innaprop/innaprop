```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Weight decay 0.0

### VGG11 - CIFAR10

```python
# Assuming data has been loaded correctly
data = pd.read_csv('best_learning_rates/cifar10/vgg11/innaprop_alpha_[0.1, 0.5, 0.9, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0]_beta_[0.1, 0.5, 0.9, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0]_epochs_20_wd_0.0/best_lr_2024-06-11_12-23.csv')

# Prepare the heatmap data
heatmap_data_loss = data.pivot("beta", "alpha", "train_loss")
heatmap_data_loss = np.log10(heatmap_data_loss)  # log scaling
heatmap_data_loss = heatmap_data_loss.iloc[::-1]  # reverse the order of rows

heatmap_data_acc = data.pivot("beta", "alpha", "test_acc")
heatmap_data_acc = heatmap_data_acc.iloc[::-1]  # reverse the order of rows

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

# Define vmin and vmax for consistent color scales
vmin_acc, vmax_acc = 75, 87 # example range for accuracy

sns.heatmap(heatmap_data_loss, ax=ax[0], annot=False, cmap="coolwarm")
ax[0].set_title('Log-scaled Training Loss', fontsize=16)
ax[0].set_xlabel('$\\alpha$',fontsize=16)
ax[0].set_ylabel('$\\beta$', fontsize=16)
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=0, fontsize=12)
ax[0].set_yticklabels(ax[0].get_yticklabels(), rotation=0, fontsize=12)
cbar = ax[0].collections[0].colorbar
cbar.ax.tick_params(labelsize=12)

# Heatmap for test accuracy
sns.heatmap(heatmap_data_acc, ax=ax[1], annot=False, cmap="coolwarm", vmin=vmin_acc, vmax=vmax_acc)
ax[1].set_title('Test Accuracy', fontsize=16)
ax[1].set_xlabel('$\\alpha$', fontsize=16)
ax[1].set_ylabel('$\\beta$',fontsize=16)
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=0, fontsize=12)
ax[1].set_yticklabels(ax[1].get_yticklabels(), rotation=0, fontsize=12)
cbar = ax[1].collections[0].colorbar
cbar.ax.tick_params(labelsize=12)


plt.tight_layout()
plt.subplots_adjust(bottom=0.2)  # Adjust bottom to accommodate the legend
plt.savefig('heatmap_vgg11_20_epochs_wd_0_0.pdf', bbox_inches='tight')
plt.show()
```
